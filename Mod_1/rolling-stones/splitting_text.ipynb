```python
text_file = open('top-500-songs.txt', 'r')
```

```python
lines = text_file.readlines()
```

```python
lines
```

```python
for i in range(len(lines)):
    lines[i] = lines[i].split('\t')
```

```python
for i in range(len(lines)):
    lines[i][-1] = lines[i][-1][:-1]
```

```python
lines
```

```python
keys = ["RANK", "NAME", "ARTIST", "YEAR"]
songlist = []
```

```python
for line in lines:
    d = {}
    for i in range(len(line)):
        d[keys[i]] = line[i]
    songlist.append(d)
```
